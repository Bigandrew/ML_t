# 回归问题
   我们先来看看回归这个词的来历：<br/>
   >“回归”是由英国著名生物学家兼统计学家高尔顿(Francis Galton,1822～1911.生物学家达尔文的表弟)在研究人类遗传问题时提出来的。为了研究父代与子代身高的关系，高尔顿搜集了1078对父亲及其儿子的身高数据。他发现这些数据的散点图大致呈直线状态，也就是说，总的趋势是父亲的身高增加时，儿子的身高也倾向于增加。但是，高尔顿对试验数据进行了深入的分析，发现了一个很有趣的现象—回归效应。因为当父亲高于平均身高时，他们的儿子身高比他更高的概率要小于比他更矮的概率；父亲矮于平均身高时，他们的儿子身高比他更矮的概率要小于比他更高的概率。它反映了一个规律，即这两种身高父亲的儿子的身高，有向他们父辈的平均身高回归的趋势。对于这个一般结论的解释是:大自然具有一种约束力，使人类身高的分布相对稳定而不产生两极分化，这就是所谓的回归效应。而回归的现代意义比起原始意义广泛许多
   * 确定响应变量与预报变量间的回归模型,即变量间相关关系的数学表达式(通常称为经验公式)
   * 根据样本估计并检验回归模型及未知参数；
   * 从众多的预报变量中，判断哪些变量对响应变量的影响是显著的，哪些是不显著的；
   * 根据预报变量的已知值或给定值来估计或预测响应变量的平均值并给出预测精度或根据响应变量的给定值来估计预报变量的值，即所谓的预报与控制问题。
回归问题与分类问题的比较：输入变量与输出变量均为连续变量的预测问题是回归问题；输出变量为有限个离散变量的预测问题成为分类问题。
   
## 线性回归
   简单的说线性回归：对于给定的数据集$D = {(x_1,y_1),(x_2,y_2),...,(x_m,y_m)}$ 其中$x_i = (x_{i1};x_{i2};...x_{id})$,$y_i\in\mathbb{R}$线性回归试图学习得$f(x_i) = \omega*x_i+b$使得$f(x_i) \simeq y_i$<br/>
如何确定$\omega$和$b$呢，均方误差$E(f;D)=\frac{1}{m}\sum_{i=1}^{n}{(f(x_i)-y_i)}^2$最小化是最常用的方法：$(\omega^*,b^*)=argmin\sum_{i=1}^{m}{(f(x_i)-y_i)}^2=argmin\sum_{i=1}{m}{(y_i-\omega*x_i-b)^2}$均方误差具有非常好的集合意义，它对应欧式几何的欧几里德距离。基于均方误差最小化的方法被称为“最小二乘法”，在线性回归中最小二乘法试图找到一条直线，使得所有样本到这条直线上的欧式距离之和最小

### 正则化
   正则化可以防止过拟合，正则化分为L1正则（lasso）和L2正则（ridge岭回归），其本质是在目标函数上加上一个关于参数绝对值和的项或者一个关于参数平方和的的扰动向即绝对值损失或者平方和损失。
   * lasso：往往会使得某些特征的参数为0，带到降维的目的，lasso可做特征感知
   * ridge：最后的拟合结果往往比lasso要好

```python
# 正则化需要加入一些超参数，超参数的取值是否合适我们使用交叉验证的方法来检验
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import GridSearchCV
data = pd.read_csv('advertising.csv')
```

```python
x = data[['TV', 'Radio', 'Newspaper']]
y = data['Sales']


x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
model = Lasso()
model1 = Ridge()

alpha_can = np.logspace(-3, 2, 10)                      #创建一个等比数列起始值为10^{-3},终值为100
lasso_model = GridSearchCV(model, param_grid={'alpha': alpha_can}, cv=5)
lasso_model.fit(x, y)
print('验证参数L1：\n', lasso_model.best_params_)


y_hat = lasso_model.predict(np.array(x_test))
mse = np.average((y_hat - np.array(y_test)) ** 2)  
print('lasso_mse:',mse)
t = np.arange(len(x_test))
plt.plot(t, y_test, 'r-', linewidth=2, label='Test')
plt.plot(t, y_hat, 'g-', linewidth=2, label='Predict')
plt.legend(loc='upper right')
plt.grid()
plt.show()

alpha_can = np.logspace(-3, 2, 10)
ridge_model = GridSearchCV(model1, param_grid={'alpha': alpha_can}, cv=5)
ridge_model.fit(x, y)
print('验证参数L2：\n', ridge_model.best_params_)
y_hat_1 = ridge_model.predict(np.array(x_test))
mse = np.average((y_hat_1 - np.array(y_test)) ** 2)  
print('ridge_mse:',mse)

t = np.arange(len(x_test))
plt.plot(t, y_test, 'r-', linewidth=2, label='Test')
plt.plot(t, y_hat_1, 'g-', linewidth=2, label='Predict')
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

## 多项式回归
   对于$f(x) = a_0+a_1x+a_2x^{2}+...+a_nx^{n}$这种形式的预测函数


```python
import numpy as np
from sklearn.linear_model import LinearRegression, RidgeCV
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
import matplotlib as mpl
```

```python
# 随机制造一些数据
if __name__ == "__main__":
    np.random.seed(0)
    N = 8
    x = np.linspace(0, 6, N) + np.random.randn(N)
    x = np.sort(x)
    y = x**2 - 4*x - 3 + np.random.randn(N)
    x.shape = -1, 1
    y.shape = -1, 1
    plt.scatter(x,y)
    plt.show()
    # 使用多项式拟合，根据范德蒙德行列，如果有6个点，那么总能找到一个5次的多项式使得6个点都在多项式的曲线上
    # 但是这样真的好么？
    model_1 = Pipeline([
            ('poly', PolynomialFeatures()),
            ('linear', LinearRegression(fit_intercept=False))])
    model_2 = Pipeline([
        ('poly', PolynomialFeatures()),
        ('linear', RidgeCV(alphas=np.logspace(-3, 2, 100), fit_intercept=False))])
    models = model_1, model_2
    mpl.rcParams['font.sans-serif'] = [u'simHei']
    mpl.rcParams['axes.unicode_minus'] = False
    np.set_printoptions(suppress=True)

    plt.figure(figsize=(9, 11), facecolor='w')
    d_pool = np.arange(1, N, 1)  # 阶数
    m = d_pool.size
    clrs = []  # 颜色
    for c in np.linspace(16711680, 255, m):
        c = int(c)
        clrs.append('#%06x' % c)   
    line_width = np.linspace(5, 2, m)
    titles = u'线性回归', u'Ridge回归'
    for t in range(2):
        model = models[t]
        plt.subplot(2, 1, t+1)
        plt.plot(x, y, 'ro', ms=10, zorder=N)
        for i, d in enumerate(d_pool):
            model.set_params(poly__degree=d)
            model.fit(x, y)
            lin = model.get_params('linear')['linear']
            if t == 0:
                print(u'%d阶，系数为：' % d, lin.coef_.ravel())
            else:
                print(u'%d阶，alpha=%.6f，系数为：' % (d, lin.alpha_), lin.coef_.ravel())
            x_hat = np.linspace(x.min(), x.max(), num=100)
            x_hat.shape = -1, 1
            y_hat = model.predict(x_hat)
            s = model.score(x, y)
            print(s, '\n')
            zorder = N - 1 if (d == 2) else 0
            plt.plot(x_hat, y_hat, color=clrs[i], lw=line_width[i], label=(u'%d阶，score=%.3f' % (d, s)), zorder=zorder)
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.title(titles[t], fontsize=16)
        plt.xlabel('X', fontsize=14)
        plt.ylabel('Y', fontsize=14)
    plt.tight_layout(1, rect=(0, 0, 1, 0.95))
    plt.suptitle(u'多项式曲线拟合', fontsize=18)
    plt.show()
```

## Logistic Regression

   在2 回归问题中我们对线性回归做了一些探讨，但是在分类问题中如何使用线性模型来进行分类呢？
   ![](1.png)<br/>
   这需要2中所提到的广义线性模型：<br/>
                                 $f(x) = g^{-1}*(w^{T}x+b)$<br/>
   其中$g^{-1}$被称为联系函数。现在我们来考虑二分类问题，其输出标记其实就两个值：0和1，而线性回归中$f(x) = w^{T}x+b$是实值，于是我们需要将实数值$f(x)$转换成0/1。这时“单位越阶函数出现了”：<br/>
                                  $y=\begin{cases}
                                      0&f(x)<0\\
                                      0.5&f(x)=0\\
                                      1&f(x)>0
                                      \end{cases}$
但该函数不连续不能直接作用于$g^{-1}$，我们需要继续需找一个函数能在一定程度上替代单位越阶函数，并且要求它是单调可微。下述内容需要一定的数学知识可跳过。，<br/>
    二分类问题转化成数学上的分布情况其实就是Bernoulli分布又叫两点分布或者0-1，如果实验成功，则伯努利随机变量取值为1，失败则为0。<br/>
    $f(x;p)=\begin{cases}
             p&x=1\\
             1-p&x=0
             \end{cases}$<br/>
上面的式子可以改写为：$f(x;p)=p^{x}*(1-p)^{1-x}$ \qquad  $x \in \{0,1\}$
统计学中有一个概念为指数族分布，满足以下形式的分布被称为指数族分布：<br/>
$p(y;\eta) = b(y)exp(\eta^{T}*T(y)-a(\eta))$<br/>
其中$T(y)$是它的充分统计量。我们指导充分统计量是不损失信息的统计量而伯努利分布也是指数族分布:<br/>
$f(x)=p^{x}(1-p)^{1-x}=e^{x\ln\frac{p}{1-p}+\ln(1-p)}$<br/>
$\eta=\ln\frac{p}{1-p}$<br/>
$e^{\eta}=\frac{p}{1-p}$<br/>
$p=\frac{1}{1+e^{-\eta}}$<br/>
其中$p=\frac{1}{1+e^{-\eta}}$是对数几率函数，我们来看看它能不能称为替代函数![](2.png)我们可以看到这个函数单调可微并一定程度上能够替代单位越阶函数，该函数是个S型函数所以它的名称是sigmoid函数，这样的回归方式又叫做对数几率回归也叫逻辑回归。<br/>
   * logistic Regression是直接对分类可能性进行建模，无需事先假设数据分布；
   * 它并不仅预测出“类别”，而是可得到近似概率的预测，这对许多需要利用概率辅助决策的任务很有帮助；
   * 对率函数是任意阶可导的凸函数，许多数值优化算法可直接用于求最优解例如：梯度下降，牛顿法等

```python
lr = Pipeline([('sc', StandardScaler()),
                        ('clf', LogisticRegression()) ])
lr.fit(x, y.ravel())
```

```python
N, M = 500, 500     # 横纵各采样多少个值

x1_min, x1_max = x[:, 0].min(), x[:, 0].max()   # 第0列的范围
x2_min, x2_max = x[:, 1].min(), x[:, 1].max()   # 第1列的范围
t1 = np.linspace(x1_min, x1_max, N)
t2 = np.linspace(x2_min, x2_max, M)
x1, x2 = np.meshgrid(t1, t2)                    # 生成网格采样点
x_test = np.stack((x1.flat, x2.flat), axis=1)   # 测试点  
cm_light = mpl.colors.ListedColormap(['#77E0A0', '#FF8080', '#A0A0FF'])
cm_dark = mpl.colors.ListedColormap(['g', 'r', 'b'])
y_hat = lr.predict(x_test)                  # 预测值
#lr.predict_proba(x_test_[0,:])             # 查看第一个样本属于各个类的概率
y_hat = y_hat.reshape(x1.shape)                 # 使之与输入的形状相同
plt.pcolormesh(x1, x2, y_hat, cmap=cm_light)     # 预测值的显示
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors='k', s=50, cmap=cm_dark)    # 样本的显示
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.xlim(x1_min, x1_max)
plt.ylim(x2_min, x2_max)
plt.grid()
plt.savefig('2.png')
plt.show()
```

```python
#看看训练集上的正确度
y_hat = lr.predict(x)
y = y.reshape(-1)
result = y_hat == y
print(y_hat)
print(result)
acc = np.mean(result)
print('准确度: %.2f%%' % (100 * acc))
```

上面的问题其实是个多分类问题而并非二分类，上面的例子中我们采用了广义线性模型softmax回归，softmax回归在深度学习中也有着广泛的应用（链接层）。其实softmax回归与逻辑回归的本质是一样的，详请参阅[softmax](http://ufldl.stanford.edu/tutorial/supervised/SoftmaxRegression/)

## 决策树
决策树（DecisionTree）是常见的一种分类学习器，一般地一颗决策树包含一个根结点，若干个内部结点和若干个叶结点，叶结点对应决策结果，其它结点则对应一个属性测试。决策树学习采用的是自顶向下的递归方法，其基本思想是以信息熵为度量构造一颗熵值下降最快的树，到叶子结点处熵值为零
要弄明白决策树我们需要搞清楚下面几个概念：
* 熵：用来衡量事情不确定性的一种指标，它应该与事件发生的可能性成负相关并且两独立事件的熵需满足可加性：$-\sum_{k=1}^{|y|}p_klogp_k$
* 信息增益: ID3算法，离散属性。
* 信息增益率：C4.5，能处理连续属性与有缺失值的属性。
* Gini系数： CART树，其实Gini系数是信息熵的近似替代。

多变量决策树：在多变量决策树中，非叶结点不再是仅对某个属性，而是对多个属性的线性组合进行测试，每个非叶结点是一个形如$\sum_{i=1}^{d}w_ia_i=t$的线性分类器，与传统的单变量决策树不同在多变量决策树中的学习过程中，不是为每个非叶结点寻找一个最优划分属性，而是试图建立一个合适的线性分类器。
在信息增益，增益率，基尼系数之外，人们还设计了不少其它的准则用于决策树的划分选择，但是对泛化性能的影响都有限。
剪枝技术是决策树学习算法对付过拟合的重要手段，分为预剪枝和后剪枝。

## 集成学习
集成学习通过构建并结合多个学习器来完成学习评估，集成学习通常会获得比单一学习器显著优越的泛化性能。要获得好的集成效果，个体学习器应该“好而不同”，如何产生“好而不同”的个体学习器，也是集成学习研究核心。目前根据个体学习器生成的方式分为：<br/>
* 个体学习器间存在强依赖关系，必须串行生成序列化的方法：Boosting。Boosting要求基学习器能够对特定的数据分布，通过“重采样”，“重赋权”的方式进行学习，从偏差-方差分解的角度来看，Boosting主要关注降低偏差。
* 个体学习器之间不存在强依赖关系，可同时生成的并行方式 Bagging。<br/>
**Random Forest**<br/>
Random Forest是Bagging的一个扩展变体，RF在以决策树为基学习器构建的Bagging集成的基础上，进一步在决策树的训练过程中引入随机属性选择，传统的决策树在选择划分时使用能使熵/Gini系数下降最快的属性，在RF中则在该结点的属性中随机选择一个包含K个属性的子集，然后再从这个子集中选择一个最优划分。<br/>
**Boosting**
提升(Boosting)是一中机器学习技术，可用于回归和分类问题，它每一步产生一个弱预测模型（如决策树）并加权累加到总模型中，如果每一步的弱预测模型生成都是依据损失函数的梯度方向，则称之为梯度提升（GBDT）；如果不是使用一阶导（梯度）信息而是使用二阶导（加速度）则构成了xgboost方法，相对于GBDT,xgboost使用二阶信息，可以更快的训练集上收敛；如果不仅仅对学习器使用权值并且对样本数据重赋权就能产生Adaboost方法

```python
import numpy as np
from sklearn import svm
import matplotlib as mpl
import matplotlib.colors
import matplotlib.pyplot as plt

def show_accuracy(a, b):
    acc = a.ravel() == b.ravel()
    print('正确率：%.2f%%' % (100*float(acc.sum()) / a.size))
```

```python
if __name__ == "__main__":
    data = np.loadtxt('14.bipartition.txt', dtype=np.float, delimiter='\t')
    x, y = np.split(data, (2, ), axis=1)
    y[y == 0] = -1
    y = y.ravel()

    # 分类器
    clfs = [svm.SVC(C=0.3, kernel='linear'),
           svm.SVC(C=10, kernel='linear'),
           svm.SVC(C=5, kernel='rbf', gamma=1),
           svm.SVC(C=5, kernel='rbf', gamma=4)]
    titles = 'Linear,C=0.3', 'Linear, C=10', 'RBF, gamma=1', 'RBF, gamma=4'

    x1_min, x1_max = x[:, 0].min(), x[:, 0].max()  # 第0列的范围
    x2_min, x2_max = x[:, 1].min(), x[:, 1].max()  # 第1列的范围
    x1, x2 = np.mgrid[x1_min:x1_max:500j, x2_min:x2_max:500j]  # 生成网格采样点
    grid_test = np.stack((x1.flat, x2.flat), axis=1)  # 测试点

    cm_light = matplotlib.colors.ListedColormap(['#77E0A0', '#FF8080'])
    cm_dark = matplotlib.colors.ListedColormap(['g', 'r'])
    matplotlib.rcParams['font.sans-serif'] = [u'SimHei']
    matplotlib.rcParams['axes.unicode_minus'] = False
    plt.figure(figsize=(10,8), facecolor='w')
    for i, clf in enumerate(clfs):
        clf.fit(x, y)

        y_hat = clf.predict(x)
        show_accuracy(y_hat, y)  # 准确率

        # 画图
        print('支撑向量的数目：', clf.n_support_)
        print('支撑向量的系数：', clf.dual_coef_)
        print('支撑向量：', clf.support_)
        print
        plt.subplot(2, 2, i+1)
        grid_hat = clf.predict(grid_test)       # 预测分类值
        grid_hat = grid_hat.reshape(x1.shape)  # 使之与输入的形状相同
        plt.pcolormesh(x1, x2, grid_hat, cmap=cm_light, alpha=0.8)
        plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors='k', s=40, cmap=cm_dark)      # 样本的显示
        plt.scatter(x[clf.support_, 0], x[clf.support_, 1], edgecolors='k', facecolors='none', s=100, marker='o')   # 支撑向量
        z = clf.decision_function(grid_test)
        z = z.reshape(x1.shape)
        plt.contour(x1, x2, z, colors=list('krk'), linestyles=['--', '-', '--'], linewidths=[1, 2, 1], levels=[-1, 0, 1])
        plt.xlim(x1_min, x1_max)
        plt.ylim(x2_min, x2_max)
        plt.title(titles[i])
        plt.grid()
    plt.suptitle(u'SVM不同参数的分类', fontsize=18)
    plt.tight_layout(2)
    plt.subplots_adjust(top=0.92)
    plt.show()
```

# 聚类

## 什么是聚类
在“无监督学习”中，训练样本的标记信息是未知的，目标是通过对无标记训练样本的学习来解释数据的内在性质以及规律，为进一步的数据分析提供基础。此类学习任务中研究最多、应用最广的是“聚类”（Clustering）。
聚类就是对大量未知标注的数据集，按数据的内在相似性将数据集划分为多个类别，使类别内的数据相似度较大而类别见的数据相似度较小，聚类属于无监督的机器学习。聚类既能作为一个单独过程，用于需找数据内在的分布结构，也可作为分类等其它学习任务的前驱过程。例如在一些商业场景中需要对新进用户进行分类判别，但在定义“用户类型”的时候可能有这样那样的困难和麻烦，此时可以对已有用户进行聚类分析，将聚类结果的每一个簇定义为一类客户，再进行分类模型的训练，对新入用户进行类别划分。

## 相似度/距离计算
   连续或有序属性的数据：<br/>
   * 闵可夫斯基距离：$dist(x_i,x_j)=(\sum_{u=1}^{n}|x_{iu}-x_{ju}|^2)^{\frac{1}{p}}$，当$p=2$时为欧式距离<br/>
   * 杰卡德相似系数(Jaccard) $J(A,B)=\frac{|A \cap B|}{|A \cup B|}$<br/>
   * 余弦相似度：$cos(\theta)=\frac{a^Tb}{|a|\cdot|b|}$<br/>
   * Pearson相似系数$\rho _{XY}=\frac{cov(X,Y)}{\sigma_X\sigma_Y}$
   * 相对熵$D(p||q)=\sum_{x}p(x)log\frac{p(x)}{q(x)}$
   * Hellinger距离 $D_{\alpha}(p||q)=\frac{2}{1-\alpha ^2}(1-\int p(x)^{\frac{1+\alpha}{2}}q(x)^{\frac{1-\alpha}{2}}dx)$<br/>
   其中余弦相似度与Pearson相似系数之间有一定的关系，相对熵与Hellinger距离之间也有一定的关系，有兴趣可以推导一下<br/>
   *离散无序的数据：VDM*
## 常见的聚类方法

   * 密度最大值聚类<br/>

### K-means
   基本思路：对于给定的类别数据K，首先给出初始划分，通过迭代改变样本和簇的隶属关系，使得每一次改进之后的划分方案都比上一次好（类别中心变化小于某个阈值）。<br/>
   K-means方法：<br/>
 
   记K个簇中心为$\mu_1,\mu_2,...,\mu_k$,每个簇的样本数目为$N_1，N_2,...N_k$使用平方误差作为目标函数：<br/>
    $J(\mu_1,\mu_2,...,\mu_k)=\frac{1}{2}\sum_{j=1}{K}\sum_{i=1}^{N_j}(x_i-\mu_j)^2$<br/>
    该函数是关于$\mu$的凸函数，驻点为$\mu_j=\frac{1}{N_j}\sum_{i=1}^{N_j}x_i$可使用梯度下降的方法求得其局部最优解，从目标函数中我们可以了解到k-means算法默认我们的样本服从高斯混合分布。*如果目标函数的平方和改为绝对值的和就变成K-mediods*<br/>
   K-means最为最广泛使用的聚类算法有着一定的局限性：
   * 将簇中所有点的均值最为新的质心，若簇中含有异常点，比如1，2，3，200的均值为51.5显然严重偏离大多数数据，这时候改进算法**K-mediods**(K-中值）就能发挥最用：取中位数3靠近大多数数据
   * K-means初值敏感：k-means方法中说明了不同的初值会取到不同的最优解，比如：<br/>
      ![](cluster1.png)<br/>
    上图中的数据比较明显的应该划分为4个绿圈里的簇但是如果随机选择的初始点如图所示，我们又不加以其它的手段进行约束进行k-means的话那么会得到以下的结果，但这并不是我们想要的结果
      ![](cluster2.png)<br/>
     这个时候我们可以使用**[K-means++](https://en.wikipedia.org/wiki/K-means%2B%2B)** <br/>
   * 由于采用批量梯度下降算法的时候需要不断的迭代求最优解对计算资源的需求比较大，尤其在处理大数据的时候。这时候不妨试试随机梯度下降，也就是mini-batch K-means,以达到更快的收敛速度<br/>
   k-means优点：<br/>
   * 简单，快速，高效的经典算法<br/>
   * 当簇为近似高斯分布时有很好的的效果<br/>
   * 一些其它聚类算法的基础(谱聚类）<br/>
   缺点：<br/>
   * 在簇的平均值可定义的情况下才能使用<br/>
   * 必须实现给出K的值(需要生成的簇的数量)<br/>
   * 不适合和发现非凸形状的簇或者大小差别很大的簇<br/>
   * 对噪声和孤立点异常敏感<br/>
   ![](cluster3.png)<br/>

```python
# K-means cluster
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
np.random.seed(0)                            #设置随机种子，用以实现实验重现
iris = datasets.load_iris()                  #sklearn自身也集成了鼎鼎大名的iris数据集
X = iris.data
y = iris.target
print(X,y)                                   #这数据熟悉吧。。
```

```python
estimators = [('k_means_iris_8', KMeans(n_clusters=8)),
              ('k_means_iris_3', KMeans(n_clusters=3)),
              ('k_means_iris_bad_init', KMeans(n_clusters=3, n_init=1,
                                               init='random'))]                         
fignum = 1
titles = ['8 clusters', '3 clusters', '3 clusters, bad initialization']
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('Petal width')
    ax.set_ylabel('Sepal length')
    ax.set_zlabel('Petal length')
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1# Plot the ground truth
fig = plt.figure(fignum, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

for name, label in [('Setosa', 0),
                    ('Versicolour', 1),
                    ('Virginica', 2)]:
    ax.text3D(X[y == label, 3].mean(),
              X[y == label, 0].mean(),
              X[y == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
ax.set_title('Ground Truth')
ax.dist = 12

fig.show()
```

```python
x = data.iloc[:,3:]               #取后两列数据方便可视化
X = x.values                       #pandas to np.array
```

```python
#使用轮廓系数查看聚类结果，轮廓系数可用于研究聚类节后簇之间的分离距离，把结果的轮廓系数可视化之后可以直观的评估簇的大小，
#轮廓系数接近+1 表示样本远离临近的簇，值为0表示该样本位于或者非常接近两个临近簇的决策边间，负值则表示这些样本可能被分配到错误的集群中
range_nclusters = [2,3,4,5,6,7]
for nclusters in range_nclusters: 
    fig,(ax1,ax2) = plt.subplots(1,2)
    fig.set_size_inches(18, 7)
    ax1.set_xlim([-0.1, 1])                              #轮廓系数取值为[-1,1]
    ax1.set_ylim([0, len(X) + (nclusters + 1) * 10])  #为每个簇提供足够的分割空间
    clusterer = KMeans(n_clusters=nclusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", nclusters,"The average silhouette_score is :", silhouette_avg)
    sample_silhouette_values = silhouette_samples(X, cluster_labels)
    y_lower = 10
    for i in range(nclusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.spectral(float(i) / nclusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),0, ith_cluster_silhouette_values,facecolor=color, 
                          edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10
        
    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")


    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
    colors = cm.spectral(cluster_labels.astype(float) / nclusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,c=colors, edgecolor='k')

    centers = clusterer.cluster_centers_   
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',c="white", alpha=1, s=200, edgecolor='k')
    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,s=50, edgecolor='k')
        
        
    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % nclusters),
                 fontsize=14, fontweight='bold')

    plt.show()
        
        
    
   
```

### 层次聚类
    层次聚类(hierarchical clustering)试图在不同层次对数据集进行划分，从而形成树形的聚类结构。它可采用“自底向上”的聚合策略，也可采用“自上而下”的分拆策略，它们分别是：
    1.聚合策略AGNES：首先将每个对象作为一个簇，然后逐渐合并。<br/>
    2.拆分策略DIANA：首先将所有对象置于一个簇中，然后逐渐细分。<br/>
    在整个聚类的过程中，往往需要计算两个类的距离来进行操作或者停止迭代的操作。判断两个类之间的距离或者说相似度常用的有以下方法：
    * SingleLinkage：取两个类中距离最近的两个样本的距离作为这两个类的距离。即最近两个样本之间的距离越小，这两个类的距离就越小。这种计算方法可能导致聚类的结果比较松散，且这种松散的效应会越来越大。比如当两个类总体上离得比较远，但却有对个别的点比较接近的情况。
    * CompleteLinkage：取两个集合中距离最远的两个点的距离作为这两个类的距离，是SingleLinkage的反面极端情况。这种计算方法可能导致两个类因为个别的点而无法合并成一个类。
    * Average-linkage：取两个类中的点两两的距离全部加在一起求平均值。这种方法得到的结果可能会受到个别点两两的距离的影响
    * Median-linkage：median-linkage是average-linkage的一个变种，取两个类中的点两两的距离全部加在一起取中值作为两个类的距离。取中值与取均值相比，更加能够减少个人偏移样本对结果的干扰。
    * Centroids：先分别计算出两个类的聚类中心，再取两个聚类中心的距离作为两个类的距离。
    * Ward’s method：假设先将两个类进行合并，得到合并后那个新cluster的均值点；再计算两个类中的每个点x到均值点的距离，再平方后加起来的结果即两个类的距离。
    
    

```python
%matplotlib inline
from sklearn.datasets.samples_generator import make_moons
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
X,y_true = make_moons(n_samples=1000,noise=0.05)
plt.scatter(X[:,0],X[:,1],c=y_true)
plt.show()
```

```python
#上图是原始图像分为两类，我们来看看层次聚类的结果
estimators = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward',compute_full_tree='false').fit(X)
plt.scatter(X[:, 0], X[:, 1], c = estimators.labels_)
plt.show()
#层次聚类和K-means一样难以发现非凸集合
```

### 谱聚类
   谱聚类是一种基于图论的聚类方法，通过对样本数据的**拉普拉斯矩阵**的特征向量进行聚类，从而达到对样本数据聚类的目的。给定一组数据$x_1,x_2,...,x_n$记任意两个点之间的相似度为$S_{ij}$，形成相似度图：$G=(V,E)$。如果$x_i$和$x_j$之间的相似度是大于一定的阈值，那么，两个点是连接的，权值记作$S_{ij}$。然后找到一个图的划分，形成若干个组，使得不同组之间有较低的权值，同一组内有较高的权值。我们来看看它的算法全貌：<br/>
   * 根据数据构造一个 Graph ，Graph 的每一个节点对应一个数据点，将相似的点连接起来，并且边的权重用于表示数据之间的相似度。把这个 Graph 用邻接矩阵的形式表示出来，记为$W$。
   * 把$W$的每一列元素都加起来得到N个数，把它们放在对角线上（其它地方都是0），变成一个N阶方阵，记为$D$。再令$L = D - W$。
   * 求出$L$的前k个特征值${\lambda}_{i=1}^k$以及对应的特征向量${v}_{i=1}^k$。
   * 把这K个特征向量排列在一起组成一个$N*K$的矩阵，将其每一行看成K维空间中的一个向量，接着使用K-means算法聚类。聚类结果每一行所属的类别就是原来Graph中的节点即最初N个数据点分别所述的类别。<br/>
   其实谱聚类就是数据降维后，在偷偷做K-means。   

### DBSCAN
   DBSCAN(Density Based Spatial Clustering of Applications with Noise)，从名字就可以看出这种聚类方法的特性：首先是基于密度的一种聚类算法，什么是密度聚类方法呢？简单的说就是样本点的密度大于某阈值，则将该样本添加到最近的簇中。密度聚类的方法能克服基于距离算法中只能发现“类圆形”（凸）聚类的缺点，可发现任意形状的簇，且对噪声数据不敏感。当然DBSACN可以发现噪声，而且DBSCAN不需要事先指定聚类的数目。有几个概念：
   * $\varepsilon$邻域：给定对象在半径 $\varepsilon$内的区域。
   * 核心对象：对于给定的数目m，如果一个对象的$\varepsilon$邻域内至少包含每个对象，则该对象为核心对象。
   * 直接密度可达：给定一个对象集合$D$,如果对象$p$是在$q$的$\varepsilon$邻域内并且$q$是一个核心对象，我们说对象$p$从对象$q$出发是直接密度可达的。
   * 密度可达：如果$q$是核心对象，$p_1$在$q$的$\varepsilon$邻域中，而$p_1$直接密度可达$p$，则$q$密度可达$p$
   * 密度相连：如果对象集合$D$中存在一个对象$x$,使得对象$p$和$q$是从$x$出发分别关于$\varepsilon$和$m$密度可达则称对象p,q是关于$\varepsilon$和$m$密度相连的。
   * 簇：基于密度的簇就是最大的密度相连对象的集合。
   * 噪声：不包含在任何簇中的对象称为噪声。<br/>
